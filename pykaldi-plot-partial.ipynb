Plotting experiments results
============================

The first cell just parse the logs, so please continue to next cells with data and graphs.

```python
import glob
from collections import namedtuple
%pylab inline
#from pylab import *

exp_names = ['mono', 'tri1', 'tri2a', 'tri2b', 'tri2b_mmi', 'tri2b_mmi_b', 'tri2b_mpe']
Experiment = namedtuple('Experiment', exp_names)


def extractResults(f):
    every_n = int(f[-6:-4])
    with open(f, 'r') as r:
        lines = r.readlines()[1:]
        assert len(lines) == 2 * \
            len(exp_names), 'Log %s should contain all experiments for zerogram and bigram' % f
        build0 = []
        for exp, line in zip(exp_names, lines[:len(exp_names)]):
            assert exp in line, 'The line in log file should start with experiments name %s' % exp
            lm, lmw, wer, ser = line.split()[2:]
            assert lm == 'build0', 'First experiments should be tested on zerogram LM'
            build0.append(wer)
        build2 = []
        for exp, line in zip(exp_names, lines[len(exp_names):]):
            assert exp in line, 'The line in log file should start with experiments name %s' % exp
            lm, lmw, wer, ser = line.split()[2:]
            assert lm == 'build2', 'First experiments should be tested on bigram LM'
            build2.append(wer)
    # TODO init namedtuple  from list, TODO iterate over namedtuple properties
    # build0_exp = Experiment(build0[0], build0[1], build0[2], build0[3], build0[4], build0[5], build0[6])
    # build2_exp = Experiment(build2[0], build2[1], build2[2], build2[3], build2[4], build2[5], build2[6])
    return (every_n, build0, build2)
```

```python
logs = glob.glob('data/*every*.log')
# TODO train data
train_data, partials = 2222, []
for f in logs:
    every_n, zerogram, bigram = extractResults(f)
    partials.append((100.0 / every_n, zerogram, bigram))
partials.sort()
data_count = [ t for t, z, b in partials]
zerogram_exps = [ z for t, z, b in partials]
bigram_exps = [ b for t, z, b in partials]
```

Graphs
======

```python
wanted = ['mono', 'tri2a', 'tri2b']
legend_list = []
for i, exp in enumerate(exp_names):
    #print len(data_count), len(zerogram_exps[i]), '\n\n'
    if exp not in wanted:
        continue
    p, = plot(data_count, [e[i] for e in zerogram_exps],'-.s')
    legend_list.append(p)
legend(legend_list, wanted, loc=3, prop={'size':'small'})
    
print('Czech generative models, zerogram')
ylabel('WER [%]')
xlabel('Portion of training size [%]')
savefig('partial-zerogram.ps', bbox_inches='tight')
```

```python
for i, exp in enumerate(exp_names):
    #print len(data_count), len(zerogram_exps[i]), '\n\n'        
    plot(data_count, [e[i] for e in bigram_exps])
ylabel('WER [%]')
xlabel('Portion of training size [%]')
print('Bigram experiments vs amount of training data')
savefig('partial-bigram.ps', bbox_inches='tight')
```
